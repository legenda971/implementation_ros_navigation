```python
import numpy as np
import matplotlib.pyplot as plt
from math import ceil


def calculate_histogram(occup_grid, sector_num):
    width, height = occup_grid.shape
    angular_resolution = 360.0 / sector_num
    
    d_max = np.sqrt((width / 2) ** 2 + (height / 2) ** 2)

    
    histogram = np.zeros(sector_num, dtype=np.float32)
    for i in range(width):
        for j in range(height):
            if occup_grid[i, j] == 0:
                continue
            
            xi = i - width / 2
            yi = j - height / 2
            
            angle = np.rad2deg(np.arctan2(yi, xi)) % 360
            bin_index = int(angle // angular_resolution)
                
            distance = np.sqrt(xi ** 2.0 + yi ** 2.0)
            influence = d_max * (1 - (distance / d_max))
            histogram[bin_index] += occup_grid[i, j]**2 * influence
    return histogram

# angles, distances = generate_lidar_data(max_distance, n_points=1)
# x, y = lidar_to_grid(angles, distances, max_distance)
# grid = create_occupancy_grid(x, y, max_distance, node_size)
```

```python
from scipy.ndimage import median_filter, gaussian_filter

def calculate_histogram(occup_grid, sector_num, threshold_factor=0.5, smoothing=5):
    width, height = occup_grid.shape
    angular_resolution = 360.0 / sector_num
    
    d_max = np.sqrt((width / 2) ** 2 + (height / 2) ** 2)
    
    histogram = np.zeros(sector_num, dtype=np.float32)
    for i in range(width):
        for j in range(height):
            if occup_grid[i, j] == 0:
                continue
            
            xi = i - width / 2
            yi = j - height / 2
            
            angle = np.rad2deg(np.arctan2(yi, xi)) % 360
            bin_index = int(angle // angular_resolution)
                
            distance = np.sqrt(xi ** 2.0 + yi ** 2.0)
            influence = (1 - (distance / d_max) ** 2)
            histogram[bin_index] += occup_grid[i, j]**2 * influence

            
    # Apply angular smoothing
    if smoothing > 0:
        histogram = np.convolve(histogram, np.ones(smoothing)/smoothing, mode='same')
            
    # Apply dynamic thresholding
    if threshold_factor > 0:
        threshold = threshold_factor * np.mean(histogram)
        histogram[histogram < threshold] = 0

    return histogram
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import ceil
import json

max_distance = 10
node_size = 1

# Function to generate random LiDAR data points
def generate_lidar_data(max_distance, n_points=100):
    angles = np.random.uniform(0, 2 * np.pi, n_points)
    distances = np.random.uniform(1, max_distance, n_points)
    return angles, distances

# Function to transform LiDAR data to grid
def lidar_to_grid(angles, distances, max_distance):
    x, y = [], []
    for angle, distance in zip(angles, distances):
        if distance == 0:
            continue
            
        x.append(distance * np.cos(angle) + max_distance)
        y.append(distance * np.sin(angle) + max_distance)
    return x, y

# Function to add noise to the LiDAR data
def add_noise(angles, distances):
    noisy_distances = distances + np.random.normal(0, distances * 0.001, len(distances))
    noisy_angles = angles + np.random.normal(0, 0.02, len(angles))
    return noisy_angles, noisy_distances

def create_occupancy_grid(x, y, max_distance, node_size=1):
    grid_size = int(ceil(max_distance * 2 / node_size)) + 1
    occupancy_grid = np.zeros((grid_size, grid_size))
    for xi, yi in zip(x, y):
        grid_x = int(xi / node_size)
        grid_y = int(yi / node_size)
        if 0 <= grid_x < grid_size and 0 <= grid_y < grid_size:
            occupancy_grid[grid_y, grid_x] += 1
    return occupancy_grid

def first_order_filter(data, alpha=0.1, adaptive=False):
    filtered_data = []
    for i in range(len(data)):
        if i == 0:
            filtered_data.append(data[i])
        else:
            # Calculate adaptive alpha
            if adaptive and i > 1:
                denominator = (data[i - 1] - data[i - 2])
                if denominator != 0:
                    alpha = min(1, abs((data[i] - data[i - 1]) / denominator))
                else:
                    alpha = 1
            print(alpha)
            filtered_data.append(alpha * data[i] + (1 - alpha) * filtered_data[i - 1])
    return filtered_data

def second_order_filter(data, alpha=0.5, beta=0.5, adaptive=False):
    filtered_data = []
    for i in range(len(data)):
        if i == 0 or i == 1:
            filtered_data.append(data[i])
        else:
            # Calculate adaptive alpha and beta
            if adaptive and i > 2:
                diff = data[i - 1] - data[i - 2]
                if diff != 0:
                    alpha = min(1, abs((data[i] - data[i - 1]) / diff))
                    beta = 1 - alpha  # beta is the remaining weight
                else:
                    alpha = 0.5  # or any default value you see fit
                    beta = 0.5  # or any default value you see fit

            filtered_data.append(alpha * data[i] + beta * data[i - 1] + (1 - alpha - beta) * filtered_data[i - 2])
    return filtered_data



```
